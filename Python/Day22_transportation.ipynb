```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd_data = pd.read_csv('내국인출국교통수단별_20230302165042.csv', encoding = 'cp949', header = 1)
pd_data.columns = ['시점', '공항', '항구']
pd_data.head()

# 공항, 항구 연 중 월별 출국 인원 plot
# normalization / standardization 적용, plot

# 2015부터 2022년까지 추이 보기.
# - 2022년 연말 공항 이용객 증가하였음.
# - 2022년 연말에 공항 출국 인원이 늘어난 것이 항구 출국인원에도 반영되었는가.
# - 연말에 출국인원이 증가하는 것이 일반적인 추이인가. --> 코로나 이전 / 이후 연도별 추이

# .sort_values()
# .sort_index()

pd_air = pd_data['공항']
pd_ship = pd_data['항구']

_, axe = plt.subplots()
axe.plot(pd_air)
axe.plot(pd_ship)
```

```python
pd_data['공항'].describe()

air_min = pd_data['공항'].min()
air_max = pd_data['공항'].max()

pd_data['공항_norm'] = pd_data['공항'].apply(lambda x: (x - air_min) / (air_max - air_min))

pd_data['항구'].describe()

ship_min = pd_data['항구'].min()
ship_max = pd_data['항구'].max()

pd_data['항구_norm'] = pd_data['항구'].apply(lambda x: (x - ship_min) / (ship_max - ship_min))

_, axe = plt.subplots()
axe.plot(pd_data['항구_norm'])
axe.plot(pd_data['공항_norm'])

print(pd_data)
```

```python
pd_data['공항'].describe()

air_mean = pd_data['공항'].mean()
air_std = pd_data['공항'].std()

pd_data['공항_std'] = pd_data['공항'].apply(lambda x: (x - air_mean) / air_std)

pd_data['항구'].describe()

ship_mean = pd_data['항구'].mean()
ship_std = pd_data['항구'].std()

pd_data['항구_std'] = pd_data['항구'].apply(lambda x: (x - ship_mean) / ship_std)

_, axe = plt.subplots()
axe.plot(pd_data['항구_norm'])
axe.plot(pd_data['공항_norm'])
axe.plot(pd_data['항구_std'])
axe.plot(pd_data['공항_std'])

print(pd_data)
```
