```python
def fact(n):
    ret = 1
    while n>1:
        ret *= n
        n -= 1
    return ret

def fact_r(n):
    if n==2:
        return n
    return fact_r(n-1)*n

def perm_f(n, r):
    return fact(n)/fact(n-r)

def combi_f(n, r):
    return fact(n)/(fact(n-r)*fact(r))

def count_with_replacement_order(pool, sel_count):  # all
    if sel_count == 1:
        return len(pool), [[x] for x in pool]
    a, b = count_with_replacement_order(pool, sel_count-1)
    r = []
    for x in pool:
        for b_i in b:
            r.append(b_i+[x])
    return len(r), r    

def perm_pool(pool, r):
    if r == 1:
        return len(pool), [[x] for x in pool]
    ret = []
    for x in pool:
        pool_copy = [y for y in pool]
        pool_copy.remove(x)
        a, b = perm_pool(pool_copy, r-1)
        for b_i in b:
            ret.append(b_i+[x])
    return len(ret), ret

def combi_pool(pool, r):
    a , b = perm_pool(pool, r)
    r = []
    for b_i in b:
        r.append(set(b_i))
    f = []
    for r_i in r:
        f.append(r_i)
        while r_i in r:
            r.remove(r_i)
    if r:
        f.append(r[0])
    return len(f), f
```

```python
import numpy as np
# 재귀 함수: 자기 자신을 부르는(사용하는) 함수
#    - 탈출 조건이 중요.
def fact_r(n):
    if n == 2:
        return n
    return fact_r(n - 1) * n


# factorial (!)
def fact(n):
    ret = 1
    while n>1:
        ret *= n
        n -= 1
    return ret

# 2 ~ n (n 포험) ndarray 작성 후 곱하기
def fact_np(n):
    return np.arange(2, n + 1).prod()
#     return np.multiply.reduce(np.arange(2, n + 1))
assert fact_np(6) == fact_r(6)
print(fact_np(40))
    
    

# 6! = 6*5*4*3*2*1 = 6*5!

def perm(n, r):
    return fact_r(n) / fact_r(n - r)

assert perm(6, 2) == 30.0
assert perm(6, 3) == 120.0

def combi(n, r):
    return perm(n, r) / fact_r(r)

assert combi(6, 2) == 15.0
```
