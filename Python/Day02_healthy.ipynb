```python
csv_data = []
with open('전국건강증진센터표준데이터.csv', encoding = 'cp949') as f:
    for line in f:
        csv_data.append(line.split(','))
print(csv_data[:3])
```

```python
name_list = []
for e in csv_data[1:]:
    if len(e[2]) > 0:
        name_list.append(e[2].split()[0])
         #print(e[2].split())
        #print(e[2].split()[0]) # e = list, e[2] = str, e[2].split = list
print(name_list)
```

```python
for n in name_list:
    if n[0] == '"':
        print(n[1:])
    else:
        print(n)
        
```

```python
for i, n in enumerate(name_list):
    name_list[i] = n.lstrip('"')
print(name_list[:10])
```

```python
#name_set = set(name_list)
for n in set(name_list):
    print(n, name_list.count(n))
```

```python
# 빈도 수 dictinory 반환
# 1. raw_list를 셋으로 변환.
# 2. 셋을 순환하면서, 각각 몇 개씩 raw_list에 있는지 카운트
# 3. 1,2에서 얻은 key, value 값으로 dictionary 만들어 return.

# lambda x: x[1]
def sort_key(x):
    return x [1]

def unique_count(raw_list, sort = False):
    raw_set = set(raw_list)
    k_list = []
    v_list = []
    
    for i in raw_set:
        k_list.append(i)
        v_list.append(raw_list.count(i))
    
    ret_dict = dict(zip(k_list, v_list))
    if sort == True:
        #ret_dict = dict(sorted(ret_dict.items(), key = lambda x: x[1]))
        ret_dict = dict(sorted(ret_dict.items(), key = sort_key))
    return ret_dict
        #print(i ,raw_list.count(i))
        
    
    
unique_count(name_list, sort = True)
```
