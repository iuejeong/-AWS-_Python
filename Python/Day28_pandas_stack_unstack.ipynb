```python
df2 = pd.read_csv('이륜차신고현황_시도별_20230310144134.csv', encoding = 'cp949', header = [0, 1, 2])
df2.drop(('시도명(1)', '시도명(1)', '시도명(1)'), axis = 1, inplace = True)
df2.columns.names = ['시점', '소유', '크기']
# print(df2.columns)
# print(df2)
# df2.stack()
# df2.stack(level = 0)
a = df2.set_index(('시군구(1)', '시군구(1)', '시군구(1)'))
b = a.stack(level = 0)
b.index.names = ['지역', '시점']
c = b.swaplevel(0, 1)
d = c.stack(level = 0)
d = d.reset_index()
```
