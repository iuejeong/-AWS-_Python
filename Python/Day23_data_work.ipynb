### 데이터 산업의 기술등급별 데이터직무 인력 현황

* 데이터 직무별 axe 생성
    - x축: 시점
    - y축: 업종 (2개)
    
    
        
* 하나의 figure에, 총 8개 axe가 바둑판 형식으로 배치되는 형태. (4 col x 2 rows), figure 사이즈는 각자 알아서.
* title 표시

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

pd_data = pd.read_csv('데이터산업의_기술등급별_데이터직무_인력_현황_20230303160506.csv', encoding = 'cp949')
pd_data.drop(columns = '기술등급별(1)', inplace = True)
pd_data.columns = ['데이터직무', '시점', '데이터 처리 및 관리 솔루션 개발·공급업', '데이터 구축 및 컨설팅 서비스업']
print(pd_data)
```

```python
list_a = [' 아키텍트', ' 개발자', ' 엔지니어', ' 분석가', '베이스 관리자', ' 과학자', ' 컨설턴트', ' 기획자']
fig, axe = plt.subplots(4, 2)
fig.set_figwidth(6.4 * 3)
for i, e in enumerate(list_a):
    filter_e = '데이터' + e
#     print(filter_e)
    filter1 = pd_data['데이터직무'] == filter_e
#     print(filter1)
    sub_data = pd_data[filter1]
#     print(sub_data)
    axes[i // 4, i % 4].plot(sub_data['시점'], sub_data.iloc[:, 2])
    axes[i // 4, i % 4].plot(sub_data['시점'], sub_data.iloc[:, 3])
    axe.set_xticks(list(np.arange(2018, 2022)))
```
