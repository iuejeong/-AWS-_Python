```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd_tips = sns.load_dataset('tips')
print(pd_tips)
```

```python
sns.catplot(data = pd_tips, x = 'day', y = 'total_bill', jitter = 0.2, hue = 'smoker', dodge = True)
sns.relplot(data = pd_tips, x= 'day', y = 'total_bill')
```

```python
sns.catplot(data = pd_tips.sort_values('day'), x = 'day', y = 'total_bill', hue = 'sex', dodge = True, kind = 'boxen')
sns.catplot(data = pd_tips.sort_values('day'), x = 'day', y = 'total_bill', hue = 'sex', dodge = True, kind = 'violin')
sns.catplot(data = pd_tips.sort_values('day'), x = 'day', y = 'total_bill', hue = 'sex', dodge = True, kind = 'violin', bw=.15, cut = 0)
sns.catplot(data = pd_tips.sort_values('day'), x = 'day', y = 'total_bill', hue = 'sex', dodge = True, kind = 'violin', split = True)
sns.catplot(data = pd_tips.sort_values('day'), x = 'day', y = 'total_bill', hue = 'sex', dodge = True, kind = 'violin', inner = 'stick', palette = "pastel")
```

```python
pd_tips.plot(x = 'day', y = 'total_bill', kind = 'line')
```
