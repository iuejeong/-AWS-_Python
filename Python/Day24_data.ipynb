### 기획:
- 어떤 내용(메시지)을 시각화 할 것인가.
    * 데이터 산업별 시장규모 성장 추이          --> line (ax.plot)
    * 2015, 2021년도 데이터 산업별 구성비 비교   ---> pie

### 설계:
- 해당 메시지를 시각화 하기 위해 어떠한 플롯을 어떻게 위치시킬 것인가.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

current_font_list = matplotlib.rcParams['font.family']

font_path = "C:\Windows\Fonts\malgunbd.ttf"
kfont = matplotlib.font_manager.FontProperties(fname = font_path).get_name()
matplotlib.rcParams['font.family'] = [kfont] + current_font_list

pd_data = pd.read_csv('데이터산업_시장규모_20230306153045.csv', encoding = 'cp949')
pd_data.loc[6, '시점'] = '2021'
pd_data.set_index(keys = ['시점'], inplace = True)
print(pd_data)
```

```python
col_names = pd_data.columns[0:]
fig = plt.figure()
a = fig.add_subplot(2, 1, 1)
b = fig.add_subplot(2, 2, 3)
c = fig.add_subplot(2, 2, 4)

for i in col_names:
    a.plot(pd_data.index, pd_data[col_names])
    b.pie(pd_data.loc['2015'])
    c.pie(pd_data.loc['2021'])
```
