```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

current_font_list = matplotlib.rcParams['font.family']

font_path = "C:\Windows\Fonts\malgunbd.ttf"
kfont = matplotlib.font_manager.FontProperties(fname = font_path).get_name()
matplotlib.rcParams['font.family'] = [kfont] + current_font_list

pd_data = pd.read_csv('데이터산업의_기술등급별_데이터직무_인력_현황_20230303160506.csv', encoding = 'cp949')
print(pd_data.head())
```

```python
fig, axes = plt.subplots(2, 4, figsize = (6.4 * 3, 4.8 * 2), sharey = True, sharex = True)
col_names = pd_data.columns[-2:]

for i, n in enumerate(pd_data['데이터직무별(1)'].unique()):
    filter1 = pd_data['데이터직무별(1)'] == n
    xs = pd_data.loc[filter1, '시점']
    ys = pd_data.loc[filter1, col_names]
    ax = axes[i // 4, i % 4]
    ax.set_title(n)
    ax.plot(xs.astype('str'), ys.iloc[:, 0], 'cv--')
    ax.plot(xs.astype('str'), ys.iloc[:, 1], 'ko--')
```
