### 전국로컬푸드인증정보표준데이터

url: https://www.data.go.kr/data/15025455/standard.do
    
    - ndarray 사용가능 (문자열 데이터 타입)

* 품목명 top 10(품목별 카운트)
    - 다중품목(+ 또는 ,로 연결)은 각각 분리
    - 이외 단일품목으로 취급.
    - 나머지는 알아서. ex) 파(실파 + 대파) --> 파 또는 [실파, 대파]
* 주소 공란: 옆칸 조회해서 메꾸기.
* dictionary 작성: {군단위 지역명: [품목명]}, 군단위 지역명 또는 기관명
* dictionary 작성: {품목 : [군단위 지역명]}

* jupyter note 업로드: 이름_로컬푸드.jpynb

```python
import numpy as np
import math

def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i])
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list

x1 = 3
y1 = 10

x2 = 5
y2 = 25

a = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
print(a)

def dist(p1, p2):      # python list ver.
    print('x:', (p2[0] - p1[0]), (p2[0] - p1[0]) ** 2)
    print('y:', (p2[1] - p1[1]), (p2[1] - p1[1]) ** 2)
    return math.sqrt((p2[0] - p1[0]) ** 2 + (p2[1] - p1[1]) ** 2)
    
def dist_np(p1, p2):     #  [3, 10]  [5, 25]
    i = p2 - p1   # [2, 15]
    i = sum(i*i)    # [4, 225]
    return math.sqrt(i)

def dist_np2(p1, p2):
    return math.sqrt(sum((p2-p1) ** 2))

    
b = dist([x1, y1], [x2, y2])   
print(b)   #15.132745950421556

c = dist_np2(np.array([x1, y1]), np.array([x2, y2]))
print(c)
```

```python
# +는 쪼개서 작성
# '군' 단위로 쪼개서 작성
# 1. 배열 순번으로 지역, 품목 나누기
# 2. 나눈 지역과 품목 안의 +들을 쪼개기
# 3. 나눈 지역을 또 OOOO도 OO군 단위로 쪼개기
# 4. 쪼갠 것들을 ndarray에 담기
# 5. 담은 것들을 dictionary 작성
csv_data = []

with open('전국로컬푸드인증정보표준데이터.csv') as f:
    for line in f:
        csv_data.append(my_split(line))
```

```python
local_data = []
for e in csv_data:
    local_data.append(e[6:8])
    
print(local_data)
```

```python
food_data = []
for i in csv_data[1:]:
    food_data.append(i[8:10])
print(food_data)
```
